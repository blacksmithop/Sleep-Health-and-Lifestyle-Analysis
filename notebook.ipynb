# Sleep Health and Lifestyle

This synthetic dataset contains sleep and cardiovascular metrics as well as lifestyle factors of close to 400 fictive persons.

The workspace is set up with one CSV file, `data.csv`, with the following columns:

- `Person ID`
- `Gender`
- `Age`
- `Occupation`
- `Sleep Duration`: Average number of hours of sleep per day
- `Quality of Sleep`: A subjective rating on a 1-10 scale
- `Physical Activity Level`: Average number of minutes the person engages in physical activity daily
- `Stress Level`: A subjective rating on a 1-10 scale
- `BMI Category`
- `Blood Pressure`: Indicated as systolic pressure over diastolic pressure
- `Heart Rate`: In beats per minute
- `Daily Steps`
- `Sleep Disorder`: One of `None`, `Insomnia` or `Sleep Apnea`

# Visualizing the data

Let us see if there is any correlation between the duration of sleep and any sleep disorders they might develop

## Correlation between Sleep Disorders and Duration of Sleep

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.barplot(x='Sleep Duration', y='Sleep Disorder', data=sleep_data) 
```

This makes sense since Insomnia is generally categorized by an inability to sleep while sleep anea is medical condition that affects the quality of sleep

```python
sns.barplot(x='Quality of Sleep', y='Sleep Disorder', data=sleep_data) 
```

## Correlation between Occupation and Sleep

```python
sns.barplot(x='Occupation', y='Quality of Sleep', data=sleep_data)
plt.xticks(rotation=45)
```

```python
sns.barplot(x='Occupation', y='Sleep Duration', data=sleep_data)
plt.xticks(rotation=45)
```

On average Engineers seem to have the longest and best quality of sleep

## Correlation between Occupation and Lifestyle choices

```python
sns.barplot(x='Occupation', y='Physical Activity Level', data=sleep_data)
plt.xticks(rotation=45)
```

```python
sns.barplot(x='Occupation', y='Stress Level', data=sleep_data)
plt.xticks(rotation=45)
```

```python
sns.barplot(x='Occupation', y='Heart Rate', data=sleep_data)
plt.xticks(rotation=45)
```

On average Sales Representatives seem to have it worse Stress and Heart Rate, this is further supported by their lowers levels of Physical Activity

## Correlation of Age and Characteristics

```python
sns.barplot(x='Age', y='BMI Category', data=sleep_data)
plt.xticks(rotation=45)
```

## Correlation between Heart Rate and Physical Activity

```python
sns.barplot(x='Heart Rate', y='Physical Activity Level', data=sleep_data)
plt.xticks(rotation=45)
```

## Visualizing as a Line Graph

```python
sns.lineplot(x="Physical Activity Level", y="Heart Rate", data=sleep_data) 
plt.show()
```

We notice that heart rate is higher for those with higher physical activity levels
